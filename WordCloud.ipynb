```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
BASE_DATA = "data/"
START_DATE = '2019-01-01'
# 1. Đọc dữ liệu
reviews = pd.read_csv(BASE_DATA+'reviews.csv', parse_dates=['review_date'])
reviews = reviews[reviews['review_date'] >= START_DATE]
reviews
```

```python
good_reviews = reviews[reviews["rating"]>3]
good_reviews
```

```python
bad_reviews = reviews[reviews["rating"]<=3]
bad_reviews
```

```python
good_dictionary = " ".join(title for title in good_reviews["review_title"])
bad_dictionary = " ".join(title for title in bad_reviews["review_title"])
```

```python
# 2. Khởi tạo WordCloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(good_dictionary)
# 3. Hiển thị hình ảnh
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off") # Ẩn trục tọa độ
plt.show()
```

```python
# 2. Khởi tạo WordCloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(bad_dictionary)
# 3. Hiển thị hình ảnh
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off") # Ẩn trục tọa độ
plt.show()
```
